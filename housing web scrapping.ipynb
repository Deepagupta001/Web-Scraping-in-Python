```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
site=requests.get("https://housing.com/in/buy/searches/AE0P38f9yfbk7p3m2h1f")
print(site)
```

```python
print(site.content)
```

```python
data = BeautifulSoup(site.content,'html.parser')
print(data.prettify())
```

```python
projectname=data.find_all('h2',class_="_gi182v _1u71grho _7s5wglyw")
print(projectname)#code nikala
```

```python
PROJECT=[]#us code ko for loop k throught list me append kiya name 
for i in data.find_all('h2',class_="_gi182v _1u71grho _7s5wglyw"):
    projectname=i.text
    PROJECT.append(projectname)
print(PROJECT)
```

```python
marketby=data.find_all('div',class_="_c81fwx _g3dlk8 _7l1v10")
print(marketby)
```

```python
MARKET=[]#us code ko for loop k throught list me append kiya name 
for i in data.find_all('div',class_="_c81fwx _g3dlk8 _7l1v10"):
    marketby=i.text
    MARKET.append(marketby)
print(MARKET)
```

```python
Price=data.find_all('div',class_="_csbfng _c8f6fq _g3gktf _ldyh40 _7l1ulh")
print(Price)
```

```python
AMOUNT=[]#us code ko for loop k throught list me append kiya name 
for i in data.find_all('div',class_="_csbfng _c8f6fq _g3gktf _ldyh40 _7l1ulh"):
    Price=i.text
    AMOUNT.append(Price)
print(AMOUNT)
```

```python
Rera=data.find_all('div',class_="_9s1o8l _vhjp4b _gzftgi _l81s5a _3f1mtk _c8sm30 _g3exct _cs1nn1 _frwh2y _2djk66 _19381tsp _1h3d1jbn _pf1rftgi _1lid1y44 _1n00oii0")
print(Rera)
```

```python
RERAAPPROVED=[]#us code ko for loop k throught list me append kiya name 
for i in data.find_all('div',class_="_9s1o8l _vhjp4b _gzftgi _l81s5a _3f1mtk _c8sm30 _g3exct _cs1nn1 _frwh2y _2djk66 _19381tsp _1h3d1jbn _pf1rftgi _1lid1y44 _1n00oii0"):
    Rera=i.text
    RERAAPPROVED.append(Rera)
print(RERAAPPROVED)
```

```python
seller=data.find_all('div',class_="css-wni7av")
print(seller)
```

```python
SALEBY=[]#us code ko for loop k throught list me append kiya name 
for i in data.find_all('div',class_="css-wni7av"):
    seller=i.text
    SALEBY.append(seller)
print(SALEBY)
```

```python
print(len(PROJECT))
print(len(MARKET))
print(len(AMOUNT))
print(len(RERAAPPROVED))
print(len(SALEBY))
```

```python
Housing=pd.DataFrame({'projectname':PROJECT,'marketby':MARKET,'Price':AMOUNT,'Rera':RERAAPPROVED,'seller':SALEBY})
Housing
```

```python
Housing.to_csv('Housing.csv',index=False)
```
